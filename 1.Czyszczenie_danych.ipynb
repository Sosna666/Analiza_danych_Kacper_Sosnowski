```python
import pandas as pd
```

```python
df = pd.read_csv("./wypadki.csv", low_memory=False) #wczytanie danych 
```

```python
#Zmieniamy nazwy kolum dla Å‚atwiejszego pisania kodu 
df.rename(columns={'ACCIDENT DATE': 'date', 'ACCIDENT TIME': 'time', 'BOROUGH': 'borough', 'ZIP CODE':'zip_code', 'LATITUDE':'latitude',
                  'LONGITUDE':'longitude','LOCATION':'location','ON STREET NAME':'on_street',
                  'CROSS STREET NAME': 'cross_street', 'OFF STREET NAME':'off_street',
                  'NUMBER OF PERSONS INJURED':'num_persons_injured', 'NUMBER OF PERSONS KILLED':'num_persons_killed',
                  'NUMBER OF PEDESTRIANS INJURED':'num_pedestrians_injured',
                   'NUMBER OF PEDESTRIANS KILLED':'num_pedestrians_killed', 
                   'NUMBER OF CYCLIST INJURED':'num_cyclist_injured',
                   'NUMBER OF CYCLIST KILLED' : 'num_cyclist_killed',
                   'NUMBER OF MOTORIST INJURED':'num_motorist_injured',
                   'NUMBER OF MOTORIST KILLED': 'num_motorist_killed',
                   'CONTRIBUTING FACTOR VEHICLE 1':'contributing_factor_vehicle_1',
                   'CONTRIBUTING FACTOR VEHICLE 2':'contributing_factor_vehicle_2',
                   'CONTRIBUTING FACTOR VEHICLE 3':'contributing_factor_vehicle_3',
                   'CONTRIBUTING FACTOR VEHICLE 4':'contributing_factor_vehicle_4',
                   'CONTRIBUTING FACTOR VEHICLE 5':'contributing_factor_vehicle_5',
                   'COLLISION_ID':'collison_id',
                   'VEHICLE TYPE CODE 1':'vehicle_type_code_1',
                   'VEHICLE TYPE CODE 2':'vehicle_type_code_2',
                   'VEHICLE TYPE CODE 3':'vehicle_type_code_3',
                   'VEHICLE TYPE CODE 4':'vehicle_type_code_4',
                   'VEHICLE TYPE CODE 5':'vehicle_type_code_5'
                  }, inplace=True)
df.head()
```

```python
#ÅºrÃ³dÅ‚o: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html
#usuwamy wiersze, ktore maja "null"
df.dropna(subset=['location'],inplace=True)
df.reset_index(drop=True, inplace=True)
df.isnull().any() 
```

```python
#Å¹rÃ³dÅ‚o: https://uszipcode.readthedocs.io/uszipcode/search.html
#Potrzebujemy jeszcze pozamieniac wartosci w kolumnie "borough" i "zip_code"
from uszipcode import SearchEngine 
search = SearchEngine()
result = search.by_coordinates(df.latitude[6],df.longitude[6])
result[0]
```

```python
print(result[4].zipcode)
print(result[4].major_city)
```

```python
#dana funcja zipcityfinder, ktÃ³ra dziÄ™ki argumentom lat i lon zwraca nam kod pocztowy i miasto podanych wspÃ³Å‚rzÄ™dnych
def WyszukiwaczMiastaIKodu(lat, lon): 
    search = SearchEngine()
    result = search.by_coordinates(lat, lon)
    return(result[0].zipcode, result[0].major_city)
```

```python
for size in range(len(df[:100])):  #stepping through 100 rows
    
    if pd.isnull(df.zip_code[size]) or pd.isnull(df.borough[size]): #Sprawdza, czy wartoÅ›Ä‡ w kolumnie zip_code lub borough jest pusta (NaN) dla danego indeksu size.
        l1,l2=df.latitude[size],df.longitude[size] #JeÅ›li tak, to pobiera wartoÅ›ci latitude i longitude dla tego samego indeksu size.
        z,c=WyszukiwaczMiastaIKodu(l1,l2) #WywoÅ‚uje funkcjÄ™ WyszukiwaczMiastaIKodu z tymi wspÃ³Å‚rzÄ™dnymi geograficznymi jako argumentami.
        df.zip_code[size]=z #Funkcja WyszukiwaczMiastaIKodu zwraca kod pocztowy (z) i nazwÄ™ gÅ‚Ã³wnego miasta (c).
        df.borough[size]=c

df.head(5)
```

```python
column_names = df.columns.tolist()
print(column_names)
```

```python
#lista kolumn, ktÃ³re chce zachowaÄ‡
other_cols = ['date', 'time', 'borough', 'zip_code', 'latitude', 'longitude', 'location', 'on_street', 'cross_street', 'off_street', 
              'num_persons_injured', 'num_persons_killed', 'num_pedestrians_injured', 'num_pedestrians_killed', 'num_cyclist_injured', 
              'num_cyclist_killed', 'num_motorist_injured', 'num_motorist_killed', 'contributing_factor_vehicle_1', 
              'contributing_factor_vehicle_2', 'contributing_factor_vehicle_3', 'contributing_factor_vehicle_4', 
              'contributing_factor_vehicle_5', 'collison_id']
```

```python
#zmieniamy sobie 5 ostatnich kolumn na 'vehicle_type' i 'vehicle_type_name'
tidy_vehicle_info_1 = df.melt(id_vars=other_cols,
                             value_vars = ['vehicle_type_code_1', 'vehicle_type_code_2', 'vehicle_type_code_3', 
              'vehicle_type_code_4', 'vehicle_type_code_5'],
                             var_name = 'vehicle_type',
                             value_name = 'vehicle_type_name') 
```

```python
#usuwamy wszystkie wiersze, w ktÃ³rych wystÄ™puje tylko 'NaN'
tidy_vehicle_info_1.dropna(subset = ['vehicle_type_name'], inplace=True) 
tidy_vehicle_info_1.reset_index(drop=True, inplace=True) 
tidy_vehicle_info_1.head() 
```

```python
#Tworzymy nowÄ… listÄ™ contributing_factors
contributing_factors = ['contributing_factor_vehicle_1', 'contributing_factor_vehicle_2',
                        'contributing_factor_vehicle_3', 'contributing_factor_vehicle_4', 
                        'contributing_factor_vehicle_5'] 
#aktualizujemy liste other_cols
other_cols = list(set(other_cols)-set(contributing_factors)) + ['vehicle_type', 'vehicle_type_name'] 
#funcja melt pozwala nam poÅ‚Ä…czyÄ‡ inne kolumny ze sobÄ… 
tidy_contributing_factor = tidy_vehicle_info_1.melt(id_vars=other_cols,
                             value_vars = contributing_factors,
                             var_name = 'contributing_factor_type',
                             value_name = 'contributing_factor_name') 
tidy_contributing_factor.head()
```

```python
#Funcja usuwa wszystkie wiersze z NaN 
tidy_contributing_factor.dropna(subset = ['contributing_factor_name'], inplace=True)
tidy_contributing_factor.reset_index(drop=True, inplace=True)
tidy_contributing_factor.head()
```

```python
#Tworzymy nowÄ… listÄ™, ktÃ³ra zawiera on_street,cross_street,off_street
street_info_cols = ['on_street','cross_street','off_street'] 
```

```python
other_cols1 = list(set(other_cols) - set(street_info_cols)) + ['contributing_factor_type','contributing_factor_name']
```

```python
id_vars_df = tidy_contributing_factor[other_cols1]
melted_dfs = []
for col in street_info_cols:
    melted_df = tidy_contributing_factor[col].reset_index(drop=True)
    melted_df['street_info'] = col 
    melted_df.columns = ['street_name', 'street_info']
    melted_dfs.append(melted_df)
value_vars_df = pd.concat(melted_dfs, ignore_index=True)
tidy_street = pd.concat([id_vars_df, value_vars_df], axis=1)
```

```python
tidy_street = tidy_street.rename(columns={0: 'street_name'})
tidy_street.head()
```

```python
tidy_street.dropna(subset = ['street_name'], inplace=True) 
tidy_street.reset_index(drop=True, inplace=True) 
tidy_street.head(50)
```

```python
tidy_street = tidy_street[tidy_street["street_name"].str.strip() != '']
tidy_street.shape
```
